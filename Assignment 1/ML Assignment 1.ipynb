```python
from sklearn.feature_extraction.text import CountVectorizer 

vectorizer = CountVectorizer(binary=True) #creating a count vectorizer object

vectorizer.fit_transform(train_text) #fit the vectorizer on the preprocessed texts

# print(vectorizer.vocabulary_) #get the vocubulary 

# transform the training and testing data using the same vectorizer
train_vector = vectorizer.transform(train_text)
test_vector = vectorizer.transform(test_text)

print(train_vector.toarray())
print()
print(test_vector.toarray())

```

# Decision Tree without Feature Selection

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(criterion = "entropy", random_state=8) #creating an instance of the decision tree object
clf.fit(train_vector, train_label)
preds = clf.predict(test_vector)
```

```python
#without feature selection
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score

print(confusion_matrix(preds, test_label))
print("The accuracy of this model is: ", accuracy_score(preds, test_label))
print("The F1 Score of this model is: ", f1_score(preds, test_label))
print("The Precision of this model is: ", precision_score(preds, test_label))
print("The Recall of this model is: ", recall_score(preds, test_label))
```
